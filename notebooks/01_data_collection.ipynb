```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup Chrome in headless mode (browser will not open)
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument("--window-size=1920,1080")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')

driver = webdriver.Chrome(options=chrome_options)

# Multiple product URLs
product_urls = [
    # Smartphones
    'https://www.amazon.in/Samsung-Galaxy-Silver-128GB-Storage/dp/B0BZCR6TNK?th=1',  # Example: Samsung Galaxy M14 5G   
  
    # Home Appliances
    'https://www.amazon.in/Mi-Vacuum-Mop-Powerful-Large-Capacity-Electronically-Controlled/dp/B0C1NYBYNQ',  # Example: Mi Robot Vacuum Mop Essential
    'https://www.amazon.in/PHILIPS-Fryer-NA120-00-Technology/dp/B0D14BB5XY/ref=asc_df_B0D14BB5XY?mcid=6762b3b95ec73ffb9c8327acc7f5b4f4&tag=googleshopdes-21&linkCode=df0&hvadid=710153233331&hvpos=&hvnetw=g&hvrand=16249532527229358096&hvpone=&hvptwo=&hvqmt=&hvdev=c&hvdvcmdl=&hvlocint=&hvlocphy=9050511&hvtargid=pla-2298444321313&gad_source=1&th=1'   # Example: Philips Air Fryer
]

data = []
for url in product_urls:
    try:
        driver.get(url)
        time.sleep(3)  # Wait for JavaScript to load

        title = driver.find_element(By.ID, 'productTitle').text.strip()
        # For price, select element safely (try multiple price elements in order)
        try:
            price = driver.find_element(By.CLASS_NAME, 'a-price-whole').text.replace(',', '')
        except:
            price = None

        data.append({
            'url': url,
            'title': title,
            'price': price,
            'date': pd.Timestamp.now()
        })
        print(f"Scraped: {title}")
    except Exception as e:
        print(f"Failed to scrape {url}: {e}")

driver.quit()

df_webscraped = pd.DataFrame(data)
print(df_webscraped)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

products = [
    'Samsung Galaxy M14 5G (ICY Silver,6GB,128GB Storage)',
    "MI Xiaomi Robot Vacuum-Mop 2i, 2200 Pa Powerful Suction, 450 mL Large-Capacity Dustbin, Electronically-Controlled 270 mL Water Tank, Controls remotely via app, Alexa/GA Enabled, White",
    "PHILIPS Air Fryer NA120/00, uses up to 90% less fat, 1500W, 4.2 Liter, with Rapid Air Technology (Black), Large"
]
urls = [
   # Smartphones
    'https://www.amazon.in/Samsung-Galaxy-Silver-128GB-Storage/dp/B0BZCR6TNK?th=1',  # Example: Samsung Galaxy M14 5G   
     
    # Home Appliances
    'https://www.amazon.in/Mi-Vacuum-Mop-Powerful-Large-Capacity-Electronically-Controlled/dp/B0C1NYBYNQ',  # Example: Mi Robot Vacuum Mop Essential
    'https://www.amazon.in/PHILIPS-Fryer-NA120-00-Technology/dp/B0D14BB5XY/ref=asc_df_B0D14BB5XY?mcid=6762b3b95ec73ffb9c8327acc7f5b4f4&tag=googleshopdes-21&linkCode=df0&hvadid=710153233331&hvpos=&hvnetw=g&hvrand=16249532527229358096&hvpone=&hvptwo=&hvqmt=&hvdev=c&hvdvcmdl=&hvlocint=&hvlocphy=9050511&hvtargid=pla-2298444321313&gad_source=1&th=1'   # Example: Philips Air Fryer

]

records = []
start_date = datetime(2025, 8, 1)
end_date = datetime(2025, 9, 3)
num_days = (end_date - start_date).days + 1

for prod, url in zip(products, urls):
    price = np.random.randint(4000, 25000)
    for day in range(num_days):
        date = start_date + timedelta(days=day)
        price += np.random.randint(-500, 500)
        records.append({
            'url': url,
            'title': prod,
            'price': abs(price),
            'date': date
        })

df_dummy = pd.DataFrame(records)
print(df_dummy)
```

```python
# Dummy data: usually already date-only, but ensure:
df_dummy['date'] = pd.to_datetime(df_dummy['date']).dt.date

# Webscraped data: timestamp to date-only 
df_webscraped['date'] = pd.to_datetime(df_webscraped['date']).dt.date
```

```python
df_webscraped['source'] = 'scraped'
df_dummy['source'] = 'dummy'
df_merged = pd.concat([df_webscraped, df_dummy], ignore_index=True)
```
